### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

import xmltodict

# 存取檔案

with open('./64_72hr_CH.xml', encoding='utf8') as fd:
    doc = dict(xmltodict.parse(fd.read()))

doc['cwbopendata']['dataset']['locations']['location']['locationName' == '鹽埕區']
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

import xmltodict

# 存取檔案

with open('./64_72hr_CH.xml', encoding='utf8') as fd:
    doc = dict(xmltodict.parse(fd.read()))
    
locs = doc['cwbopendata']['dataset']['locations']['location']
for loc in locs:
    for area in loc['weatherElement']:
        if area['description'] == '溫度':
            print(loc['locationName'], area['time'][0]['dataTime'], area['time'][0]['elementValue']['value'])
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

import xmltodict

# 存取檔案

with open('./64_72hr_CH.xml', encoding='utf8') as fd:
    doc = dict(xmltodict.parse(fd.read()))
    
loc = doc['cwbopendata']['dataset']['locations']['location'][0]
print(loc['locationName'])
for i in loc['weatherElement']:
    if i['description'] == '溫度':
        for j in i['time']:
            print(i['time'][0]['dataTime'], j['elementValue']['value'])
```
