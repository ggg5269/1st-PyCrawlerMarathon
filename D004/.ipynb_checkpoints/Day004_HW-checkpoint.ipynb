### 自行尋找一個合適的 API 接口做練習，並且查看其回傳內容

* https://cat-fact.herokuapp.com/facts (來源：https://alexwohlbruck.github.io/cat-facts/)
* http://odata.wra.gov.tw/v4/RealtimeWaterLevel (來源：https://data.gov.tw/dataset/25768)


```python
r3 = requests.get("https://pokeapi.co/api/v2/berry/1")
r3.text
json.loads(r3.text)
```

```python
import pandas as pd
poke = pd.DataFrame()
temp = []

berry = requests.get("https://pokeapi.co/api/v2/berry/")
Len = len(json.loads(berry.text)['results'])

for i in range(Len):
    temp.append(json.loads(berry.text)['results'][i]['name'])
    if i == Len - 1:
        poke['berries'] = temp
```

```python
firmness = []
flavors = []
growth = []
maxHarvest = []
NaturalGiftPow = []
NaturalGiftType = []
size = []
smoothness = []
dryness = []
for i in range(1, Len + 1):
    attr = requests.get("https://pokeapi.co/api/v2/berry/{}/".format(i))
    jattr = json.loads(attr.text)

    firmness.append(jattr['firmness']['name'])
    flavors.append([(jattr['flavors'][i]['flavor']['name'] + ": " + str(jattr['flavors'][i]['potency']))  for i in range(len(jattr['flavors']))])
    growth.append(jattr['growth_time'])
    maxHarvest.append(jattr['max_harvest'])
    NaturalGiftPow.append(jattr['natural_gift_power'])
    NaturalGiftType.append(jattr['natural_gift_type']['name'])
    size.append(jattr['size'])
    smoothness.append(jattr['smoothness'])
    dryness.append(jattr['soil_dryness'])
    if i == Len:
        poke['firmness'] = firmness
        poke['flavors'] = flavors
        poke['growth'] = growth
        poke['maxHarvest'] = maxHarvest
        poke['NaturalGiftPow'] = NaturalGiftPow
        poke['NaturalGiftType'] = NaturalGiftType
        poke['size'] = size
        poke['smoothness'] = smoothness
        poke['dryness'] = dryness
```
