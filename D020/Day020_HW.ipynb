```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
items = table.find_all('tr')[1:]
```

```python
d = {}
for i in items:
    if(len(i.find_all('td'))) == 5:
        param = i.find_all('td')[0].get_text()
        d[param] = {}
        month = i.find_all('td')[2].get_text()
        d[param][month] = i.find_all('td')[3].get_text()
    else:
        date = i.find_all('td')[0].get_text()
        d[param][date] = i.find_all('td')[1].get_text()
d
```
