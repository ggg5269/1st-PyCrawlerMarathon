```python
import requests
r = requests.get('https://www.etmall.com.tw/')
response = r.text

response
```

### 用上述網站說明該如何判斷 Header 中必須加上哪些資料才會正確


```python
headers = {'user-agent': 'my-app/0.0.1'}
r = requests.get('https://www.etmall.com.tw/',headers=headers)
response = r.text

response
```
